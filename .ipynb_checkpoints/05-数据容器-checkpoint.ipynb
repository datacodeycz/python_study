````python
# 创建列表操作
print([])
a = []
print(a)
a = list()
print(a)

# 创建列表时，就给a容器存放数据
a = [1,2,3,4,5]
print(a)
# a = list(1,2,3,4,5)  # 会报错

# list() 主要用处，是将其他的数据容器转换成列表容器
# list() 创建容器，只能创建空的列表。
````

### 列表的操作


1.访问列表的中的元素<br>
```列表名[num]```
注意：表中存储的每一个元素都有对应的索引，我们可以通过索引去访问对应的元素 索引是 开始到末尾是从0开始到length-1，而末尾到开始是从-1到-length<br>
2.修改列表中对应的元素-- 通过下标去修改列表中对应的元素<br>
 ``列表名[num] = 需要修改的值``<br>
3.新增元素<br>
(1)```列表名.insert(参数1，参数2)``` 在指定的下标位置，插入指定的元素<br>
(2)```列表名.append(元素)``` 将指定元素，增加到列表的尾部<br>
(3)```列表名.extend(容器)```将参数容器的内容依次取出，插入到列表的尾部<br>
4.删除元素<br>
(1)``del 列表名[下标]``<br>
(2)`` 列表名.pop[下标]``<br>
(3)```列表.remove(元素)``` 从正方向，删除出现的第一个元素<br>
5.清空元素<br>
```列表名.clear()``` 清空列表<br>
6.统计元素个数<br>
```len(列表名)```<br>
7.通过元素值查找元素对应的下标--注意返回的是正下标<br>
```列表名.index(元素值)```<br>
8.统计某个元素在列表中出现的次数<br>
```列表名.count```<br>

### 列表的遍历
1.使用while循环，遍历列表中的元素<br>
<br>
2.使用for循环，遍历列表中的元素<br>

````python
# while
list1 = [3,1,4,2,5,3,2]

index = 0
while index < len(list1):
    print(list1[index]);
    index = index + 1

````

````python
# 使用while循环遍历
list1 = [3,1,4,2,5,3,2]

# index为下标
index = 0
while index < len(list1):
    # 在控制台输出下标为index的元素
    print(list1[index])
    # 控制index往后 + 1 
    index = index + 1
````

````python
# 使用for循环遍历
list2 = [2,3,123,12,43,4]
for x in list2:
    # 输出x从list2中获取到的元素
    print(x)
````

小结：<br>
1.使用while循环可以人为的控制条件并输出的结果，例如 我可以控制输出下标为奇数对应的元素，只需要将 index = 1  index = index + 2即可<br>
2.使用for循环，更简单，更直观的体现了遍历的过程，并且只能将所有元素进行遍历，但是这不妨碍for循环控制输出的结果，可以借助if语句和continue关键字<br>

## 数据容器-tuple(元组)
有时候我们需要一个容器来存放重要的数据，这个重要的数据是不允许被修改的，如果我们还用列表的话，列表中的元素是会被修改的，所以我们要引入一个新的数据容器-元组，**元组中的元素一旦定义确定后，是不可被修改的**

### 元组的创建
定义元组使用小括号，并且使用逗号隔开各个数据，数据**可以是不同数据类型**

````python
# 创建空元组
()  # 空元组

print(tuple())  # 空元组

a = ()

# 创建时给元组增加元素

a = (1,2,3,4)
print(a)

a= tuple([1,2,3])  # 用于创建空元组,可以将其他的数据容器中的数据存储到元组中
print(a)

# 通过list() 将元组中的元素存放到列表

b = list((1,2,3,4))
print(b)
````

````python
# 元组的创建
# 1.定义一个没有变量指向的字面量
(1,2,3,4)

# 用一个变量指向元组这个字面量
a = ()
a = ("a","b",3,4)
print(a)
````

````python
# 2. 使用tuple()函数创建空元组容器

b = tuple()  # tuple()函数主要是将其他数据容器转换成元组容器，所以如果直接通过基本数据类型字面量的方式是无法将元素写入到元组中的
print(a)
print(b)
````

### 元组的操作
由于元组在创建的时候就已经确定的元组中的元素，所以无法进行增删改的操作，只能进行查找操作<br>
操作有<br>
1.元组名\[num\]:访问元组中对应下标的元素值<br>
2.index():查找某个元素，返回对应的下标，如果不存在会报错<br>
3.count()：统计某个元素在当前元组中出现的次数<br>
4.len(元组):统计元组内的元素个数<br>

````python
tuple_test = (1,2,3,4)
print(tuple_test[2])

tuple_test.index(3)

````

````python
tuple_test = (1,1,1,2,4,5)
print(tuple_test.count(1))
len(tuple_test)
````

````python
test_tuple = (1,2,3,4,5,1,1)

# 访问元素值
print(test_tuple[1])
# 1.index()
print(test_tuple.index(2))
# 2.count()
print(test_tuple.count(1))
# len()
print(len(test_tuple))
````

### 元组的遍历
while循环语句<br>
for循环语句<br>

````python
# while循环语句
index = 0
while index < len(test_tuple):
    print(test_tuple[index])
    index = index + 1
````

````python
# for循环遍历
for x in test_tuple:
    print(x)
````

### 元组小结
1.可以存储多个数据<br>
2.可以存放不同数据类型的数据<br>
3.元组中的数据都是有对应的下标的<br>
4.允许出现重复的数据<br>
5.**不可以被修改**<br>
6.可以使用for循环遍历查找<br>

### 字符串的操作
字符串跟之前学过的列表和元组一样，都有元素对应的下标，可以通过元素对应的下标去找到对应的元素，这里的字符串存储的主要是一个个字符，可以理解成将一个字符串拆分成了多个字符，以一个容器的形式进行存储

操作:<br>
1.通过下标访问对应的字符<br>
2.str.index():通过给定字符返回第一个匹配的下标<br>
3.str.replace(字符串1，字符串2):将字符串内的全部字符串1中的字符串，全部替换为字符串2，该操作是非破坏性，会得到一个新的字符串<br>
4.str.split(字符串):按照指定的分隔符将字符串划分为多个字符串，并存入一个新的列表对象中，进行返回，非破坏性<br>
5.str.strip(字符串):移除首尾指定的字符串，该操作是非破坏性的操作，会以一个新的字符串进行返回；注意：如果没有参数，表示移除空格字符,<br>
6.str.count(字符串):返回该参数字符串出现的次数<br>
7.len(字符串)：返回字符串的字符个数<br>

````python
# 访问指定下标的字符
test_str = "yuanchengzhi"
test_str[3]

# 通过指定的字符访问对应的下标的第一个
test_str.index('n')

# 字符串中的指定部分替换为其他字符串
# print(test_str.replace("cheng",""))
# print(test_str)
# 按照指定的分隔符进行分隔
print(test_str.split("n"))

print(test_str.count('n'))
print(len(test_str))



# 应用
# input返回值：我们控制台输入的内容，该内容是一个字符串
# a  = input("请输入你的java，js，mysql的成绩(成绩之间用逗号隔开):")
# java,js,mysql = a.split(",")
# print(java,js,mysql)
````

### 字符串的扩展
format()函数，我们可以使用该函数格式化字符串，可以用于print()的输出问题

````python
# 实例
# 格式如下
name = "yuanchengzhi"
age = "18"
address = "四川遂宁"
money = 100.5
# {} 表示 占位符
# 1. 正常情况下，第一个{}匹配第一个变量，依次类推，匹配多个变量
"我的名字是：{}，我今年{}岁了，我家在{}".format(name,age,address)  # 可以通过该方法灵活的书写字符串中的信息，将字符串进行格式化
# 2.可以在{}中填写信息来指定匹配的变量是哪一个，在format()方法中，第一个变量表示0
"我的名字是{0},我今年{1}岁了，再次说明我的名字是{0}".format(name,age)  # 使用数字的方式进行匹配
# 3.{}中该可以指定要匹配变量的格式，例如你想让money匹配的结果是一个浮点数
"我身上有{:.2f}块".format(money)  # 表示输出的格式为小数点后两位，注意这里只是改变输出的格式，并没有改变原变量
````


`^`,` <`, `>` 分别是居中、左对齐、右对齐，后面带宽度， : 号后面带填充的字符，只能是一个字符，不指定则默认是用空格填充。

`+` 表示在正数前显示 `+`，负数前显示 `-`；  （空格）表示在正数前加空格

`b`、`d`、`o`、`x` 分别是二进制、十进制、八进制、十六进制。

## 序列的切片操作（非常重要）
序列是指：内容连续、有序、可使用下标索引的一类数据容器，就是我们上述学到的列表、元组、字符串、均可以视为序列<br>
序列的常用操作-切片<br>
切片的意思是：在原序列中切出一个子序列<br>
切片的语法格式是：`序列[起始下标:结束下标:步长]`<br>
`起始下标`：开始分割（切片）的起点，表示从这个位置开始，包含这个元素。不填表示下标为0的为起点<br>
`结束下标`：结束分割（切片）的终点，表示从这个位置结束，但不包括这个元素。不填默认到最后一个元素，但包含最后一个元素<br>
`步长`：元素与元素之间的差值。不填默认为1<br>


````python
# 实例代码
# 列表序列
test_list = [1,2,3,4,5,6]
new_test_list = test_list[1:2]
new_test_list1 = test_list[1::2]
new_test_list1
````

````python
# 实例代码
# 元组序列
test_tuple = (1,2,3,4,5,6)
new_test_tuple = test_tuple[1:-1]
new_test_tuple1 = test_tuple[1:-1:2]
print(new_test_tuple,new_test_tuple1)
````

````python
# 实例代码
# 字符串
test_str = "abcdef"
new_test_str = test_str[1:-1]
new_test_str1 = test_str[1:-1:2]
print(new_test_str,new_test_str1)
````

## 数据容器-set(集合)
通过上述学习到的三个数据容器，我们能看出来，它们都可以支持重复元素的，但是有时候我们的需求，是希望一个容器里面存放唯一的元素，就是没有重复的元素，这个数据容器就是set集合，在set集合中，是不允许出现重复元素的

### 集合的创建
集合的创建主要是借助{}实现的，{}中存放元素，元素之间用逗号分开

````python
# 实例代码
# 集合的创建
# 创建一个字面量的集合
{1,2,3,4}
{}  # 空的集合字面量
# 创建一个用变量指向的集合
test_set = {1,2,3,4}
test_set1 = set()  # 该方式只能将其他数据容器的类型转换成集合，不能用于创建有数集合，只能创建一个空集合
print(test_set)
print(test_set1)
````

````python
# 测试 - 在创建集合的过程中 如果出现了重复的元素， 会发生什么
a = {1,2,2,1}
print(a)  # 通过结果能看出来，该集合中的元素直接将重复的元素去除掉了，并没有报错
````

### 集合的操作
1.由于集合是无序的（元素之间没有对应的索引），所以集合是不支持下标索引访问的<br>
2.集合和列表一样，里面的元素是允许被修改的<br>
<br>
修改的一些操作<br>
1.添加指定元素`set.add(元素)`，如果集合中存在该元素，不会保证，正常执行，只是不会添加进去<br>
2.删除指定元素，`set.remove(元素)`，,返回值为None，就是说明只是单纯的删了，如果集合中不存在该元素，则删除失败 并报错<br>
3.从集合中随机取出元素`set.pop()` 从集合中随机取出一个元素，这个元素会在该集合中被删除，需要一个变量去接收这个被取出的元素<br>
4.清空集合`set.clear()`<br>
5.获取两个集合的差集：`集合1.difference(集合2)`<br>
&emsp;功能：取出集合1和集合2的差集（集合1中有的 集合2中没有的）<br>
&emsp;输入（参数）：1.要进行差集的集合<br>
&emsp;输出（返回）：返回一个进行差集的结果，集合1和集合2不变<br>
6.删除2个集合中的差集：`集合1.difference_update(集合2)`<br>
&emsp;功能：删除集合1中和集合2相同的元素<br>
&emsp;输入：进行比较的集合<br>
&emsp;输出：None<br>
7.获取两个集合的并集`集合1.union(集合2)`<br>
&emsp;功能：集合1和集合2合并成新的集合<br>
&emsp;输入：需要合并的集合<br>
&emsp;输出：返回合并后的集合，原集合不变<br>
访问的一些操作<br>
1.`len(set)`:统计集合内有多少元素<br>

````python
# 实例练习
test_set = {1,2,3,4}
# 添加指定元素add()
test_set.add(1)  # 在集合中存在了，正常执行，但不会添加进去了
test_set.add(5)  # 将5添加进去了
print(test_set)  
````

````python
# 删除指定元素
print(test_set.remove(1))  # 删除一个集合中存在的元素
print(test_set)
# test_set.remove(6)  # 删除一个集合中不存在的元素 和删除失败并报错
````

````python
# 随机取出元素
test_set = {1,2,3,4}
set_num = test_set.pop()
print(set_num)
print(test_set)

# 问题：如果集合中没有元素了，怎么办 还能取出来吗
test_set1 = {}
# test_set1.pop()  # 当集合中没有元素后，如果再向集合中取元素，会报错
````

````python
# 清空集合
test_set2 = {1,2,3,4,5}
test_set2.clear()  # 清空集合中的元素
print(test_set2)
````

````python
# set.difference()
test_set1 = {1,2,3,4}
test_set2 = {1,2,5,6}
new_test_set = test_set1.difference(test_set2)
print(new_test_set)  # set1中有的，set2中没有的
````

````python
# set.difference_update()
test_set1 = {1,2,3,4}
test_set2 = {1,2,5,6}
test_set1.difference_update(test_set2)  # 删除了集合1中与集合2中重复的元素，集合2不变
print(test_set1,test_set2) 
````

````python
# set.union()
test_set1 = {1,2,3,4}
test_set2 = {1,2,5,6}
new_test_set = test_set1.union(test_set2)
print(new_test_set)
print(test_set1)
print(test_set2)

# 上述代码表示进行了合并，但是并没有去改变集合1和集合2中的元素
````

````python
# len()
test_set1 = {1,2,3,4,5}
print(len(test_set1))
````

### 集合的遍历
由于for循环遍历是从容器中取元素，所以我们可以使用for循环进行遍历。<br>
但是while循环遍历是我们借助索引下标进行遍历，集合没有索引下标，所以不能直接实现遍历，但是我们可不可以间接实现遍历啦？

````python
# for循环遍历
test_set_for = {"a","b","y","l"};
for x in test_set_for:
    print(x,end="")
````

````python
# while循环遍历  高级

test_set_while = {1,2,3,4}
while True:
    if(len(test_set_while) == 0):
        break
    print(test_set_while.pop())
````

### 集合小结
1.集合中可以容器多个数据<br>
2.集合中可以容纳不同类型的数据<br>
3.数据是无序存储的（不支持下标索引）<br>
4.集合中的元素都是唯一的，没有重复元素<br>

## 数据容器-dict(字典)
字典这里容器，我们可以通过生活中的字典来讲述，字典中有很多字，每一个字都应对着不同的意思，“字”跟“字的意思”是一一对应的关系，同样在我们字典容器一样，我们有一个字”对应着“字的意思”，就比如说我们的成绩，你的学号或是姓名，对应着你的成绩，都是一一对应的，虽然说我们的成绩可能一致，但是这个成绩属于不同学生的成绩。<br>

### 字典的创建
字典的概念：<br>
&emsp;字典的存储方式是以键值对的方式存储的，key-value
字典的定义：<br>
&emsp;使用{}，存储的元素是键值对（一对一对的形式）

````python
# 实例
test_dict = {"ycz":100,"qwe":98,"asd":90}
print(test_dict)

'''
ps:
1.key - value 可以是任意类型的数据，但是key不能是一个字典
2.key不可重复，重复会对原数据覆盖
'''
test_dict1 = dict()  # 使用dict()函数可以创建一个空的字典
print(test_dict1)
````

### 字典的操作
说明：由于字典也没有索引下标，所以不能使用下标索引来获得对应的值，但是可以通过key来找到对应的value值<br>
1.使用key找到对应的value `dict[key]`<br>
2.新增元素`dict[字典中不存在的key] = value`  注意：key不可重复，所以如果使用了字典中存在的key，会将之前的key对应的value覆盖掉<br>
3.删除元素`dict.pop(key)`通过key找到对应的value，删除该键值对<br>
4.清空元素`dict.clear()` 字典被修改，元素被清空<br>
5.获取字典中所有的key值 `dict.keys()`,返回字典中的全部key<br>
6.获取字典中全部元素的数量`len(dict)`<br>

````python
# 实例代码
# 1.使用key找到对应的value  这里的key就很像之前学习的列表的下标，只是列表的下标只能是0，1 这种， 这里的key可以是任意值（除了字典）
test_dict = {
    "yuanchengzhi":{
        "java":100,
        "mysql":100,
        "js":100
    }
}
test_dict["yuanchengzhi"]
````

````python
# 增加一个新的元素
test_dict = {
    "yuanchengzhi":{
        "java":100,
        "mysql":100,
        "js":100
    }
}
test_dict["qwe"] = {
    "java":90,
    "mysql":89,
    "js":90
}
print(test_dict)
````

````python
# 删除元素
test_dict = {
    'yuanchengzhi': {
        'java': 100, 
        'mysql': 100, 
        'js': 100
    }, 
    'qwe': {
        'java': 90, 
        'mysql': 89, 
        'js': 90
    }
}
test_dict.pop("qwe");
print(test_dict)
````

````python
# 清空字典中的元素
test_dict = {"yuanchengzhi":{"age" : 18}}
test_dict.clear()
print(test_dict) # 清空字典中的元素
````

````python
# 计算字典内的元素数量（注意：判断哪个字典就计算的哪个字典内元素的数量，只有字典中的字典数量不在考虑范围内）
test_dict = {"yuanchengzhi":{"age" : 18}}
print(len(test_dict));  # 1
````

### 字典的遍历
字典的遍历一般主要是借助key去遍历，通过key去遍历value，或是直接遍历key</br>

````python
# 使用for循环进行遍历
test_dict = {
    "01": "qwe",
    "02": "asd",
    "03": "zxc"
}

# 下面的方式x获取的是字典中key，可以通过这种方式来遍历key，然后通过key找到value
for x in test_dict:
    '''遍历key'''
    print(x)

for x in test_dict:
    '''遍历key:value'''
    print(f"{x}:{test_dict[x]}")

'''
注意：由于只能通过key来找到value，而在不知道key值的时候，是无法找到key-value的，并且key的遍历只能通过for循环，
所以字典不能使用while来实现
'''
````

### 字典的特点
1. 可以容纳多个数据
2. 可以容纳不同类型的数据（注意key不能是字典类型的）
3. 每一份数据都是key-value键值对类型的
4. 可以通过key来获取vlaue，且key不可重复（当输入了重复的key，会覆盖之前的value）
5. 不支持下标索引
6. 可以修改（增加或删除修改元素）
7. 可以通过for循环遍历，但是不能通过while循环遍历

## 数据容器的通用功能
1. 都可以通过for循环实现遍历
2. max()方法 返回容器中的最大元素
3. min()方法 返回容器中的最小元素
4. len()方法 返回容器元素的个数
5. list()方法 转换为列表
6. tuple()方法 转换为元组
7. str()方法 转换为字符串
8. set()方法 转换为集合
9. sorted(序列，\[reverse = True])  排序，reverse = Ture为默认值，表示降序，返回结果为排序后的**列表**
</br>
注意:容器之间相互转换是非常常用的手段，有很多技巧，可以帮助使用

````python
# 上述的的max() min() len()方法都以列表为例子讲解
test_list = [1,2,3,4,5]
test_list1 = ["a","z","b",1]

# max()
print(max(test_list))  # 得到结果为 5

# min()
print(min(test_list))  # 得到结果为 1

# len()
print(len(test_list))  # 得到数量为5
````

````python
# 当要判断最大值和最小值的数据容器(除字符串外)中有非数字类型，还可以判断吗
test_list = ["a","z","b",1]
# max(test_list)  # 报错
# min(test_list)  # 报错

# 当数据容器是字符串的时候，不会报错，会按照ASCII码进行判断大小，字符都有对应的数字，
# 将字符按照ASCII码转换为数字然后再判断大小 
test_str = "helloworld"
print(max(test_str))
print(min(test_str))
````

````python
# 扩展：字符串的比较
# 字符串的比较是按位比较，也就是一位一位的比较，只要有一位大，纳闷整体大
test_str1 = "abc"
test_str2 = "abd"
print(test_str1 > test_str2)  # False  c < d
````
