### 双分支语句
双分支语句，意味着有两种结果，而这两种结果是判断该条件产生的结果，一个条件要不是真，要不是假，不会出现第三种情况，
所以当该条件为真时，应该做什么，当该条件为假的时候，应该做什么。<br>


````python

# 实例
# 当你的成绩在60分及以上时，你不需要参加补考，并且能度过一个好过的暑假，如果在60分以下，那就需要补考，并且需要在暑假的时候，好好准备补考

score = int(input("请输入你的成绩："))
if score >= 60:
    # 当成绩为60分及以上时，执行该代码块
    print("恭喜你，你的成绩在60分以上，所以你不需要参加补考，希望你能度过一个愉快的暑假")
else: # 使用else关键字，表示 不满足上述条件最后的结果，else的意思是 其他
    # 当成绩为60分以下的时候，执行该代码块
    print("很可惜，你的成绩在60以下，所以你需要参加补考，希望你能在暑假的时候好好准备补考，一定会通过补考")
````

### 多分支语句
多分支语句，意味着会有多个结果，而产生多个结果需要我们判断多个条件才能得到多个结果，如果只有一个条件，最多只能产生两个结果，而现在需要产生多个结果，就需要多个条件;<br>
<br>
ps：多个条件之间可以没有任何关系，但是如果条件判断的是同一个问题就需要注意，在书写条件的时候，条件之间一定不能有交集，如果有交集会导致程序在某种情况下出错

````python
# 实例
# 通过你输入的成绩，判断你成绩的等级，90-100 为A，80-90 为B，70-80 为 C，60-70为D，60分以下为E
score = int(input("请输入你的成绩："))
if score > 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("E")
    
# 问题：这里我并没有将条件完整的写出来，为什么还是能实现该功能
# 隐含条件问题
````

##  判断语句的嵌套
在实际中，我们需要在满足前置条件后再进行判断，例如如果我们要判断一个同学是否挂科，有一种方式：首先我们得判断他的java成绩是否挂科，如果他Java成绩没有挂科，那我们再判断他的js成绩是否挂科，如果js成绩没有挂科，那需要判断他的mysql成绩是否挂科，如果mysql也没有挂科，说明该同学没有挂科。
这种需求，我们可以通过判断语句的嵌套来实现，就是在判断第一个条件，满足第一个条件后再判断第二个条件，依次类推

````python
# 实例
# 实现上述功能
score_java = int(input("请输入你的Java成绩："))
score_js = int(input("请输入你的js成绩："))
score_mysql = int(input("请输入你的mysql成绩："))
if score_java >= 60:
    if score_js >= 60:
        if score_mysql >= 60:
            print("恭喜你，你的所有成绩都及格了")
        else:
            print("抱歉，你有没及格的科目")
    else:
        print("抱歉，你有没及格的科目")
else:
    print("抱歉，你有没及格的科目")
````

## while循环
while循环的基本结构<br>
<br>
```
while 条件:
    条件为真时，执行的代码块
    （代码块执行完，就再次判断条件)
```
<br>
ps：1.当条件一直为真时，会无限循环执行，而无限循环并不是我们想要的，所以就提出一个问题<br>
问题1：怎么写才能使我们不执行无限循环，换句话说，我们需要设置终止的条件，保证不无限执行<br>
答：当我们的需求是执行50次循环，或是找到一个值时就结束循环，只有将我们的代码块跟条件关联的时候，才能不执行无限循环。<br>

````python
# 实例while 流水线的工人，从早上9点开始要工作到12点，每一个小时进行记录
# time 表示时间
time = 9
while time < 12:
    print("工人们正在工作")
    time = time + 1
````

````python
# 练习1 在100-999中找一个三位数，满足其各位数字的立方和确好等于该数本身
i = 100
while  i < 999:
    h = int(i / 100)
    ten = int(i / 10 % 10)
    u = int(i % 10)
    result = h ** 3 + ten ** 3 + u ** 3
#     print(result)
    if( result == i):
        print(i)
    i = i + 1
````

````python
# 练习2 求1-100的和，并输出
# 存储我们计算的结果
sum = 0
# 进行循环的标识，作为控制我们循环次数的一个标识
i = 1
while i <= 100:
    sum = sum + i
    i = i + 1
print(sum)
print("====================")
# 练习3 在1-20找到偶数，并输出
j = 1
while j <= 20:
    if j % 2 == 0:
        print(j)
    j = j + 1
````

````python
# 进阶练习：设置一个1-100的随机数，然后猜这个随机数，没猜对重新猜，猜对结束
# 第一步
import random

# 设置一个1-100的随机数
# (x,y)
# n * (y - x) + x
random_1_100 = random.randint(1,101) # 获得了一个1-100随机整数
while True:
    num = int(input("请输入一个1-100的随机数:"))
    if random_1_100 > num:
        print("不好意思,你猜小了")
    elif random_1_100 < num:
        print("不好意思,你猜大了")
    else:
        print("恭喜你猜对了")
        break
````

### while循环语句的嵌套使用基本格式
在需求设计中，有时候我们会有这样的需求，在这一个月里，流水线的工人都需要在9-12点之间工作，并且每一个小时会检查一次

````python
# while循环的嵌套的案例

day = 1
while day <= 30:
    # 工作天数
    time = 9
    while time < 12:
        # 时间
        print(f"day{day},{time}点工人正在正常工作")
        time = time + 1 
    day = day + 1
````

````python
# 练习  九九乘法表
i = 1
while i <= 9:
    j = 1
    while j <= i:
        print(f"{j} * {i} = {i * j}\t",end="")
        j = j + 1
    print("\t")
    i = i + 1
````

## for循环
for循环的结构，
```
for 临时变量 in 待处理的数据集:
    循环执行的代码块
```
python中的for循环更多的是作为的一个循环遍历的功能，在其他语言中，for循环可以像while循环一样去使用，但是在python中更多的是去遍历一个容器的元素

````python
# 实例

# 这里的字符串，就是像JavaScript中的一个数组，首先按照这种方式理解，后续会继续讲解
str_1 = "math-model"
for x in str_1:
    print(x,end="")
````

通过上述代码我们能看出来，for循环更多的实现的是一个遍历容器元素的操作 所有我们需要学到后面才会讲解，这里我们主要通过range()函数生成一个简单的数字序列进行讲解<br>
range()<br>
&emsp; 功能：生成一个数字序列，生成的范围由参数决定<br>
&emsp; 输入（参数）：<br>
&emsp; &emsp; 参数1：（1）当只有参数1的时候，获取一个从0开始，到参数1结束的数字序列；（2）当有两个参数时，参数1表示起始位置<br>
&emsp; &emsp; 参数2：表示数字序列中的终点位置（不包括参数2）<br>
&emsp; &emsp; 参数3：表示数字之间的步长，未输入参数3时，默认为1<br>
&emsp; &emsp;  步长：数字与数字之间的间隔，例如1和3之间的步长就是2<br>
&emsp;输出（返回值）：返回上述范围的数字序列<br>
<br>
注意：range()返回的是一个可迭代对象（类型是对象），而不是列表类型， 所以打印的时候不会打印列表

````python
# range()函数的使用
sum = range(5)  # 表示[0,5)之间的数字序列
print(sum)
for x in sum:
    print(f"{x} ",end="")
````

````python
# 两个参数的时候
sum = range(1,10) # 1-9
for x in sum:
    print(f"{x} ",end="")
````

````python
# 三个参数的时候
for x in range(1,10,2):
    print(f"{x} ",end="")
````

小插曲：python中的作用域
在python中首先目前为止我们只需要探讨全局作用域和局部作用域
在Python中，只有模块，类以及函数才会引入新的作用域，像我们上述代码中写的x，在其他语言中，i是在for循环这个块作用域的，我们在全局中是无法使用这个变量的，但是在python中它仍然处于当前的作用域，我们外部是可以使用x这个变量

````python
# 作用域测试
for x in range(1,5):
    print(x)
    
print(x)  # 4 就是访问到数字序列中的最后一个元素
# 在正常开发过程中，我们按照编程规范来表示，在一个代码块中的变量，我们的外层就尽量不要使用
````

### for循环的嵌套使用
for循环中的嵌套使用其实跟while循环的逻辑是一样的，在for循环嵌套使用中我们得从线变成面

````python
# 在这一个月里，流水线的工人都需要在9-12点之间工作，并且每一个小时会检查一次
for day in range(1,31):
    for time in range(9,12):
        print(f"day{day},{time}点工人正在正常工作")
````

````python
# 练习：使用for循环打印九九乘法表
for i in range(1,10):
    for j in range(1,i + 1):
        print(f"{j} * {i} = {i * j}\t",end="",sep="\t")
    print("\t")
````

## 循环中断
有时候我们也有这样的需求，当我们去循环查找一个元素的时候，当找到这个元素的时候，我们就不需要继续执行该循环，就需要中断循环，而中断循环又会出现两种情况：<br>
（1）流水线上的某个工人在某一天请假的，他就未正常工作，但是他其他时间都正常工作了。（跳出某一次循环） continue<br>
（2）流水线上的某个工人在干到第28天的时候，辞职不干了，后面都不会来了。（跳出整个循环）break<br>

### break
退出整个循环  相当于辞职了，后面的工作都不干了

````python
# 实例练习
# 在3-100中找一个数，这个数是一个偶数，当我们找到第一个偶数就退出

for x in range(3,101):
    if(x % 2 == 0):
        print(f"第一个偶数是{x}")
        break
````

### continue 
退出某一次循环，相当于有一天，请假了，后面的工作还是要继续的干

````python
for x in range(1,21):
    if(x % 2 != 0):
        continue
    print(x)
````

````python
# 实例练习
# 在1-20中找到偶数，并且偶数输出，要求使用continue
for x in range(1,21):
    if(x % 2 != 0):
        continue
    print(f"{x}")
````

````python
# 注意
# 当程序中出现了break，break后面即使有代码 也不会执行
for x in range(1,101):
    if(x == 27):
        break
        print(x)
    print(x)
````

小结：<br>
```break```可以退出整个循环<br> 
```continue```只能退出某一次循环，在使用过程中，一般情况下我们会结合着分支语句进行使用<br>
