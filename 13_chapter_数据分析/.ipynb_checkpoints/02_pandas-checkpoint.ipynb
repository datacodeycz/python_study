## 使用 Pandas 读取 CSV 文件

[Pandas](https://pandas.pydata.org/)是一个流行的 Python 库，用于处理表格数据（类似于电子表格中存储的数据）。Pandas 提供了从 CSV、Excel 电子表格、HTML 表格、JSON、SQL 等各种文件格式读取数据的辅助函数。让我们下载一个文件`italy-covid-daywise.txt`，其中包含以下格式的意大利 Covid-19 日数据：

```

2020-04-21,2256.0,454.0,28095.0
2020-04-22,2729.0,534.0,44248.0
2020-04-23,3370.0,437.0,37083.0
```

这种数据存储格式被称为 **逗号分隔值** 或 CSV。

> **CSV**： 逗号分隔值（CSV）文件是一种使用逗号分隔数值的分隔文本文件。文件的每一行都是一条数据记录。每条记录由一个或多个字段组成，字段之间用逗号隔开。CSV 文件通常以纯文本格式存储表格数据（数字和文本），在这种情况下，每一行都有相同数量的字段。(维基百科）

## 从数据帧中获取数据

你可能想做的第一件事就是从数据帧中获取数据，例如，特定日期的计数或特定列中的值列表。要做到这一点，了解数据帧中数据的内部表示可能会有所帮助。从概念上讲，你可以把数据帧看作一个**列表字典**：键是列名，值是包含各列数据的列表/数组。

````python
# Pandas format is simliar to this
covid_data_dict = {
    'date':       ['2020-08-30', '2020-08-31', '2020-09-01', '2020-09-02', '2020-09-03'],
    'new_cases':  [1444, 1365, 996, 975, 1326],
    'new_deaths': [1, 4, 6, 8, 6],
    'new_tests': [53541, 42583, 54395, None, None]
}

covid_data_dict['date']
````

以上述格式表示数据有几个好处：

* 一列中的所有值通常具有相同的值类型，因此将它们存储在一个数组中更有效率。
* 检索特定行的值只需从每列数组中提取给定索引处的元素。
* 与其他为每行数据使用字典的格式相比，这种表示法更紧凑（列名只记录一次）（见下面的示例）。

````python
# Pandas format is not similar to this
covid_data_list = [
    {'date': '2020-08-30', 'new_cases': 1444, 'new_deaths': 1, 'new_tests': 53541},
    {'date': '2020-08-31', 'new_cases': 1365, 'new_deaths': 4, 'new_tests': 42583},
    {'date': '2020-09-01', 'new_cases': 996, 'new_deaths': 6, 'new_tests': 54395},
    {'date': '2020-09-02', 'new_cases': 975, 'new_deaths': 8 },
    {'date': '2020-09-03', 'new_cases': 1326, 'new_deaths': 6},
]
````

````python
covid_data_list[0]['new_cases']
````

有了列表字典的类比，你现在就可以猜测如何从数据帧中获取数据了。例如，我们可以使用"[]"索引符号从特定列中获取值列表。

## 分析数据帧中的数据

让我们试着回答一些有关数据的问题。

**问：意大利报告的与Covid-19相关的病例和死亡总数是多少？**

与 Numpy 数组类似，Pandas 数列也支持使用 "sum "方法来回答这些问题。

````python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
````

````python
print('The number of reported cases is {} and the number of reported deaths is {}.'.format(int(total_cases), int(total_deaths)))
````

**问：总死亡率（报告的死亡人数与报告的病例数之比）是多少？**

````python
death_rate = covid_df.new_deaths.sum() / covid_df.new_cases.sum()
````

````python
print("The overall reported death rate in Italy is {:.2f} %.".format(death_rate*100))
````

**问：总共进行了多少次测试？在报告每日测试次数之前，共进行了 935310 次测试。**


````python
initial_tests = 935310
total_tests = initial_tests + covid_df.new_tests.sum()
````

````python
total_tests
````

**问：检测结果呈阳性的比例是多少？**

````python
positive_rate = total_cases / total_tests
````

````python
print('{:.2f}% of tests in Italy led to a positive diagnosis.'.format(positive_rate*100))
````

## 使用日期

虽然我们已经查看了病例、测试、阳性率等方面的总体数据，但按月研究这些数据也很有用。在这里，"日期 "列可能会派上用场，因为 Pandas 提供了许多处理日期的实用程序。

````python
covid_df.date
````

date 的数据类型目前是 `object`，因此 Pandas 并不知道该列是日期。我们可以使用 `pd.to_datetime` 方法将其转换为 `datetime` 列。

````python
covid_df['date'] = pd.to_datetime(covid_df.date)
````

````python
covid_df['date']
````

可以看到，它现在的数据类型是 `datetime64`。现在，我们可以使用 `DatetimeIndex` 类（[查看文档](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DatetimeIndex.html)）将数据的不同部分提取到单独的列中。

````python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
````

````python
covid_df
````

让我们检查一下五月份的总体指标。我们可以查询五月份的行，选择列的子集，然后使用 `sum` 方法汇总每个选定列的值。

````python
# Query the rows for May
covid_df_may = covid_df[covid_df.month == 5]

# Extract the subset of columns to be aggregated
covid_df_may_metrics = covid_df_may[['new_cases', 'new_deaths', 'new_tests']]

# Get the column-wise sum
covid_may_totals = covid_df_may_metrics.sum()
````

````python
covid_may_totals
````

````python
type(covid_may_totals)
````

我们还可以将上述操作合并为一条语句。

````python
covid_df_sum = covid_df[covid_df.month == 5][['new_cases', 'new_deaths', 'new_tests']].sum()

covid_df_sum
type(covid_df_sum)
````

## 分组和汇总

下一步，我们可能想汇总按日划分的数据，并创建一个包含按月划分的数据的新数据帧。我们可以使用 `groupby` 函数为每个月创建一个组，选择要汇总的列，并使用 `sum` 方法对它们进行汇总。

````python
covid_month_df = covid_df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()
````

````python
covid_month_df
````

其结果是一个新的数据帧，该数据帧使用传递给 `groupby` 的列中的唯一值作为索引。分组和聚合是一种强大的方法，可将数据逐步汇总到更小的数据帧中。

除了按总和聚合外，还可以按平均值等其他指标聚合。让我们计算一下每个月的日均新增病例数、死亡数和检验数。

````python
covid_month_mean_df = covid_df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].mean()
````

每个月的`new_cases`, `new_deaths`, `new_tests`数据的平均值

````python
covid_month_mean_df
````

除分组外，另一种聚合形式是截至每一行日期的病例、测试或死亡的运行或累计总和。我们可以使用 `cumsum` 方法将一列的累计和作为一个新系列来计算。让我们添加三个新列： 病例总数"、"死亡总数 "和 "检测总数"。

````python
covid_df['total_cases'] = covid_df.new_cases.cumsum()
````

````python
covid_df['total_deaths'] = covid_df.new_deaths.cumsum()
````

````python
covid_df['total_tests'] = covid_df.new_tests.cumsum() + initial_tests
````

我们还在`total_test`中加入了初始测试次数，以反映每日报告开始前进行的测试。

````python
covid_df
````

请注意，"total_tests "列中的 "NaN "值并没有受到影响。

我们可以通过添加更多列将这些数据合并到现有的数据帧中。不过，要合并两个数据帧，我们至少需要一个共同的列。让我们在 `covid_df` 数据框中插入一个 `location` 列，并将所有值都设为 `Italy`。

````python
covid_df['location'] = "Italy"
````

````python
covid_df
````

现在，我们可以使用 `.merge` 方法将 `locations_df` 中的列添加到 `covid_df` 中。

````python
merged_df = covid_df.merge(locations_df, on="location")
````

````python
merged_df
````

意大利的位置数据被附加到`covid_df`中的每一行。如果 `covid_df` 数据框包含多个地点的数据，那么每一行都会附加相应国家的地点数据。

现在，我们可以计算每百万人中的病例数、每百万人中的死亡数和每百万人中的测试数等指标。

````python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
````

````python
merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population
````

````python
merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population
````

````python
merged_df
````
