# 数据容器

在学习一个新的概念时，我们通常需要回答以下三个问题：

1. **为什么要用数据容器**  
   在程序开发中，往往需要将多个数据组织在一起，例如存储一个学生的多项信息（姓名、年龄、成绩等）。如果只是单纯依靠字面量操作，显然无法满足需求。此时，我们需要一个容器来存放这些数据，并方便地管理和使用它们。

2. **数据容器是什么**  
   在 Python 中，数据容器是一种可以容纳多份数据的数据结构，用于存储和操作数据。数据容器根据其特性可以存储不同类型的数据，提供不同的操作方法。

3. **数据容器怎么用**  
   学完本章后，你将能够回答如何创建和操作数据容器的问题。

## 数据容器的类型
Python 提供了多种数据容器，每种容器都有其独特的特性和用途：
- **列表（list）**：有序、可变的数据容器。
- **元组（tuple）**：有序、不可变的数据容器。
- **字符串（str）**：表示文本数据的容器，本质上是字符的有序集合。
- **集合（set）**：无序、不重复的容器。
- **字典（dict）**：键值对存储数据的容器。

每种容器根据其特性适用于不同的场景，例如：
- 列表和元组适合存储有序的数据。
- 集合适合去重操作。
- 字典适合用来快速查找和存储键值对数据。
